```python
import pandas as pd
df = pd.read_csv('Automobile_data.csv', sep=',')
```

```python
df
```

```python
#checking null values
dfresult=df.dropna()
print(dfresult)
df.isnull().sum()
```

```python
#checking duplicates
print(df.duplicated())
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
df.dtypes
df
```

```python
df=df.drop(df[df.normalized_losses.str.contains(r'[?]')].index)
df=df.drop(df[df.bore.str.contains(r'[?]')].index)
df=df.drop(df[df.stroke.str.contains(r'[?]')].index)
df=df.drop(df[df.horsepower.str.contains(r'[?]')].index)
df=df.drop(df[df.peak_rpm.str.contains(r'[?]')].index)
```

```python
df[['normalized_losses','bore','stroke','horsepower','peak_rpm','price']] = df[['normalized_losses','bore','stroke','horsepower','peak_rpm','price']].apply(pd.to_numeric)
df[['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','fuel-system']]=df[['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','fuel-system']].astype("string")
```
